```python
from IPython.display import display
from PIL import Image, ImageDraw, ImageColor
def draw_point(image: Image.Image, correct_point: list, wrong_point : list, wrong_point_1: list,color=None):
    if isinstance(color, str):
        try:
            color = ImageColor.getrgb(color)
            color = color + (128,)  
        except ValueError:
            color = (255, 0, 0, 128)  
    else:
        color = (255, 0, 0, 128)  
    red_color = (255, 0, 0, 128)
    green_color = (0, 255, 0, 128)
    dark_red_color = (128, 0, 0, 128)
    dark_green_color = (0, 128, 0, 128)
    overlay = Image.new('RGBA', image.size, (255, 255, 255, 0))
    overlay_draw = ImageDraw.Draw(overlay)
    radius = min(image.size) * 0.05
    if correct_point:
        x, y = correct_point

        overlay_draw.ellipse(
            [(x - radius, y - radius), (x + radius, y + radius)],
            fill=dark_green_color,
        )
        
        center_radius = radius * 0.15
        overlay_draw.ellipse(
            [(x - center_radius, y - center_radius), 
            (x + center_radius, y + center_radius)],
            fill=green_color
        )
    if wrong_point:
        x, y = wrong_point
        overlay_draw.ellipse(
            [(x - radius, y - radius), (x + radius, y + radius)],
            fill=dark_red_color,
        )
        
        center_radius = radius * 0.15
        overlay_draw.ellipse(
            [(x - center_radius, y - center_radius), 
            (x + center_radius, y + center_radius)],
            fill=red_color
        )
    if wrong_point_1:
        x, y = wrong_point_1
        overlay_draw.ellipse(
            [(x - radius, y - radius), (x + radius, y + radius)],
            fill=dark_red_color,
        )
        
        center_radius = radius * 0.15
        overlay_draw.ellipse(
            [(x - center_radius, y - center_radius), 
            (x + center_radius, y + center_radius)],
            fill=red_color
        )
    image = image.convert('RGBA')
    combined = Image.alpha_composite(image, overlay)

    return combined.convert('RGB')   
def display_image(image_path: Image.Image,correct_point=None,wrong_point=None,wrong_pint_1= None, title=None):
    image = Image.open(image_path)
    image_point = draw_point(image, correct_point,wrong_point,wrong_pint_1)
    save_path = 'output.jpg'
    image_point.save(save_path)
    display(image_point)
```
